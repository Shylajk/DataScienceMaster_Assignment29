<H5> Importing Libraries</H5>

```python
import numpy as np
from scipy import misc
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

<H4> Download the Raw image of Racoon in grey scale </H4>

```python
face = sp.misc.face(gray = True)
plt.figure(figsize=(8, 3.5))
plt.subplot(121)
plt.imshow(face, cmap='gray')
plt.show()
```

<H4> Compressing the Racoon image </H4>

```python
# Reshaping the image
rows = face.shape[0]
cols= face.shape[1]
image  = face.reshape(rows*cols,1)
```

<H4> Compressing the grey image with KMeans with 5 clusters </H4>

```python
n_clusters = 5
kmeans = KMeans(n_clusters = n_clusters, n_init=4, max_iter=10)
kmeans.fit(image)

clusters = np.asarray(kmeans.cluster_centers_,dtype=np.uint8) 
labels = np.asarray(kmeans.labels_,dtype=np.uint8 )  
labels = labels.reshape(rows,cols); 

plt.imsave('new raccon.png',labels)
image = plt.imread("new raccon.png")
plt.figure(figsize=(8, 3.5))
plt.subplot(121)
plt.imshow(image)
plt.show()
```
